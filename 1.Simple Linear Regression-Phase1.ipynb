```python
import pandas as pd #Pandas is a library where we can handle data-related tasks, and 'as pd' means alias name for pandas
```

```python
dataset=pd.read_csv("Salary_Data.csv") #read the input file and store/assign in to one variable 
```

```python
dataset #the variable is having input dataset (full structure) and simply we can run the assigned variable, it will show whatever is being held
```

```python
independent=dataset[["YearsExperience"]] #independent variable means it holds input values from dataset
```

```python
independent #(display the input value) simply we can run the assigned variable, it will show whatever is being held
```

```python
dependent=dataset[["Salary"]] #dependent variable means it holds output values from dataset
```

```python
dependent #(display the output value) simply we can run the assigned variable, it will show whatever is being held
```

```python
#Splitting the train and test data by using the train test split function from sklearn.
#And here assigning the independent and dependent data set respectively to the train and test variable. 
#also used test_size and randam stage for splitting the datas according to the given limits.
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,y_train)
'''the linearregression function is used for fit the input and output values which means used to create the model'''
```

```python
weight=regressor.coef_
weight
'''coef is used to get the weight value from train model'''
```

```python
bais=regressor.intercept_
bais
'''intercept is used to get the bais value from train model'''
```

```python
y_pred=regressor.predict(X_test)
#predict function is used for predict the value from input test dataset for validation purpose (becz already we have output test dataset for comparison)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
#as mentioned earlier, we can validate between trained model(predicted) with test output
```

```python
r_score ##'''r_score is helped to get the difference value between test and predicted value, if we get lesser than 1 score values , then we can 
##consider this is a good model else it is a bad model(need to recheck) '''
```

```python
import pickle #pickle function is used to store the model and output check from trained model
filename1="finalized_model_linear.sav"
```

```python
pickle.dump(regressor,open(filename1,'wb')) # dump means store the trained model with write binary mode
```

```python
loaded_model=pickle.load(open("finalized_model_linear.sav",'rb')) # load means loaded the trained model file with read binary mode
```

```python
result=loaded_model.predict([[15]]) #this is final step which means we can see the linear regression output using by predict function
```
